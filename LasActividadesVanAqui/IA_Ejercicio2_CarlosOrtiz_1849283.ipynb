EJERCICIO 2

Crear un juego de piedra, papel o tijera de dos jugadores. El programa preguntará por la opción del jugador 1 y la opción del jugador 2 y dará como resultado quien ganó el juego. 

```python
#ADQUISICIÓN DE DATOS 
print ("Elije uno: 1.Piedra, 2.Papel, 3.Tijera")
J1= int (input("Jugador 1, coloque su elección: ")) 
J2= int (input("Jugador 2, coloque su elección: "))
#PROCESO 
if J1 == 1: 
  J1 = "piedra" #Se iguala que el número 1 es igual a piedra 
elif J1 == 2:
  J1 = "papel"  #Se iguala que el número 2 es igual a papel 
elif J1 == 3:
  J1 = "tijera" #Se iguala que el número 3 es igual a tijera 

if J2 == 1:
  J2 = "piedra" #Se iguala que el número 1 es igual a piedra
elif J2 == 2:
  J2 = "papel"  #Se iguala que el número 2 es igual a papel
elif J2 == 3:
  J2 = "tijera" #Se iguala que el número 3 es igual a tijera

if J1 == "piedra" and J2 == "papel" : 
  print("El ganador es el jugador 2") #Se indica que el papel le gana a la piedra y el ganador es el J2
elif J1 == "piedra" and J2 == "tijera" :
  print("El ganador es el jugador 1") #Se indica que la piedra le gana a la tijera y el ganador es el J1
elif J1 == "papel" and J2 == "piedra" :
  print("El ganador es el jugador 1") #Se indica que el papel le gana a la piedra y el ganador es el J1
elif J1 == "papel" and J2 == "tijera" :
  print("El ganador es el jugador 2") #Se indica que la tijera le gana al papel y el ganador es el J2
elif J1 == "tijera" and J2 == "piedra" :
  print("El ganador es el jugador 2") #Se indica que la piedra le gana a la tijera y el ganador es el J2
elif J1 == "tijera" and J2 == "papel" :
  print("El ganador es el jugador 1") #Se indica que la tijera le gana al papel y el ganador es el J1
elif J1 == J2 :
  print("Empate")  #Se indica que si ambo eligen lo mismo es empate
```
