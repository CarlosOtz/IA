Ejercicio N°2. Crear un programa que le pida al usuario capturar dos números enteros y dos números flotantes. Después, realizará la suma, resta, multiplicación y división, de ambos pares de números. Como resultado se deberán de obtener 8 valores. Imprimir todos estos reultados.

```python
#ADQUISICIÓN DE DATOS 
x1 = int(input("Ingrese el primer número entero: "))#Se le pide al usuario un número entero 
x2 = int(input("Ingrese el segundo número entero: "))#Se le pide al usuario un segundo número entero 
y1 = float(input("Ingrese el primer número flotante: "))#Se le pide al usuario un número flotante 
y2 = float(input("Ingrese el segundo número flotante: "))#Se le pide al usuario un segundo número flotante
#ENTEROS
se = (x1+x2)#Se le asigna la suma de los valores enteros 
re = (x1-x2)#Se le asigna la resta de los valores enteros 
me = (x1*x2)#Se le asigna la multiplicación de los valores enteros 
de = (x1/x2)#Se le asigna la división de los valores enteros 
#FLOTANTES 
sf = (y1+y2)#Se le asigna la suma de los valores flotantes
rf = (y1-y2)#Se le asigna la resta de los valores flotantes
mf = (y1*y2)#Se le asigna la multiplicación de los valores flotantes
df = (y1/y2)#Se le asigna la división de los valores flotantes
#RESULTADOS 
print("La suma de los números enteros es: ",se)
print("La resta de los números enteros es: ",re)
print("La multiplicación de los números enteros es: ",me)
print("La división de los números enteros es: ",de)
print("La suma de los números flotantes es: ",sf)
print("La resta de los números flotantes es: ",rf)
print("La multiplicación de los números flotantes es: ",mf)
print("La división de los números flotantes es: ",df)
```
